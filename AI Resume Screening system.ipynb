```python
from transformers import pipeline

hf_pipeline = pipeline(
    "text-generation",
    model="gpt2"
)

def llm(prompt):
    result = hf_pipeline(
        prompt,
        max_new_tokens=80,   # ✅ correct
        do_sample=False
    )
    return result[0]['generated_text']

# Test
print(llm("AI is"))
```

```python
def extract_prompt(resume):
    return f"""
Extract only:
Skills:
Tools:
Experience:

Resume:
{resume}

Answer in short.
"""

def match_prompt(resume_data, job_description):
    return f"""
Find:

Matched Skills:
Missing Skills:

Resume:
{resume_data}

Job:
{job_description}

Give only answer.
"""

def score_prompt(text):
    return f"""
Give only a number between 0 to 100.

{text}
"""

def explanation_prompt(score, text):
    return f"""
Explain in 2 lines:

Score: {score}

{text}
"""
```

```python
def run_pipeline(resume):

    # Step 1: Extract (basic)
    extracted = resume.lower()

    # Step 2: Matching (manual logic)
    job_skills = ["python", "machine learning", "nlp", "sql", "tensorflow", "powerbi"]

    matched = []
    missing = []

    for skill in job_skills:
        if skill in extracted:
            matched.append(skill)
        else:
            missing.append(skill)

    # Step 3: Scoring (manual)
    score = int((len(matched) / len(job_skills)) * 100)

    # Step 4: Explanation
    explanation = f"""
Matched Skills: {matched}
Missing Skills: {missing}
Candidate has {len(matched)} out of {len(job_skills)} required skills.
"""

    return score, explanation
```
